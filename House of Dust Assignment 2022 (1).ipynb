```python
import emoji
#!/usr/bin/python

# A House of Dust, copyright (c) 2014 Nick Montfort <nickm@nickm.com>
# Original by Alison Knowles & James Tenney, 1967
#
# Permission to use, copy, modify, and/or distribute this software for any
# purpose with or without fee is hereby granted, provided that the above
# copyright notice and this permission notice appear in all copies.
#
# THE SOFTWARE IS PROVIDED "AS IS" AND THE AUTHOR DISCLAIMS ALL WARRANTIES
# WITH REGARD TO THIS SOFTWARE INCLUDING ALL IMPLIED WARRANTIES OF
# MERCHANTABILITY AND FITNESS. IN NO EVENT SHALL THE AUTHOR BE LIABLE FOR ANY
# SPECIAL, DIRECT, INDIRECT, OR CONSEQUENTIAL DAMAGES OR ANY DAMAGES
# WHATSOEVER RESULTING FROM LOSS OF USE, DATA OR PROFITS, WHETHER IN AN
# ACTION OF CONTRACT, NEGLIGENCE OR OTHER TORTIOUS ACTION, ARISING OUT OF OR
# IN CONNECTION WITH THE USE OR PERFORMANCE OF THIS SOFTWARE.
#
# Updated 10 March 2015 to remove a duplicate value in "place".
# Updated 17 November 2015 to remove a near-duplicate value in "inhabitants".
# Updated 31 May 2018 to add compatibility with Python 3 (Python 2 still works)

from random import choice
import time

material = [':zap:', ':bamboo:',':cloud:', ':fallen_leaf:', ':cherry_blossom:', ':sparkles:', \
            ':mushroom:', ':ear_of_rice:', ':crescent_moon:', ':star:', ':fire:', ':snowflake:', ':musical_note:', \
           ':thought_balloon:', ':shell:']

location = ['IN A GREEN, MOSSY TERRAIN :mountain_bicyclist:', 'IN A :city_sunrise:', 'BY THE SEA :palm_tree:', 'IN A DESERTED FACTORY :office:', \
            ':evergreen_tree: IN DENSE WOODS :deciduous_tree:', 'BY A :house_with_garden:', 'IN :japan:', 'BY A :circus_tent:', 'BY MOUNT FUJI :mount_fuji:', \
            'IN A COLD, WINDY CLIMATE :dash:', 'IN A PLACE WITH HEAVY RAIN :umbrella:', 'ON THE SEA :ship:',':house: AMONG OTHER HOUSES :house:', \
            'IN A DESERTED CHURCH :church:', 'IN A METROPOLIS :city_sunset:', 'BY A :hotel:', 'AT A BRIDGE AT NIGHT :bridge_at_night:', 'BY A :school:']

light_source = ['A FLASHLIGHT :flashlight:', 'JACK O LANTERNS :jack_o_lantern:', 'ALL AVAILABLE LIGHTING :high_brightness:', 'ELECTRICITY :bulb:', \
                'THE :sunny: LIGHT', 'THE :full_moon: LIGHT', ':fire: LIGHT', 'LOW LIGHT :low_brightness:']


inhabitants = [ 'PEOPLE WHO EAT :ramen:','PEOPLE WHO DRIVE :taxi:',':man: :older_woman: :girl: :girl:',\
               'MEN :man:', 'A COUPLE :couple_with_heart:', 'OLDER WOMEN :older_woman:', ':horse: AND :bird:',\
               'ANGELS :angel:', 'PERFORMERS :performing_arts:', ':baby: AND :older_man:', 'PEOPLE WHO BIKE :bike:', \
               ':octopus: :fish: :tropical_fish:', 'ALIENS :alien:', 'PEOPLE WHO ENJOY SINGING :notes:', 'PRINCESSES :princess:', \
               'DANCERS :dancer:', ':boy: AND :girl:', ':two_women_holding_hands:', 'PEOPLE WHO OWN :cat:', ':dog:', 'FAMILIES :family:', \
               'PEOPLE WHO LOVE TO READ  :books: :blue_book:', ':honeybee: :honeybee: :honeybee:', ':ghost:']

while True:
    print('')
    print(emoji.emojize('A HOUSE OF ' + choice(material), use_aliases=True))
    print(emoji.emojize(('      ' + choice(location)),use_aliases=True))
    print(emoji.emojize('            USING ' + choice(light_source),use_aliases=True))
    print(emoji.emojize('                  INHABITED BY ' + choice(inhabitants),use_aliases=True))
    print('')
    time.sleep(8)
```

&nbsp;&nbsp; For this project, I approached combinatorial poetry with many ideas. I decided to use the Emoji PyPi library to incorporate the use of emojis in the project. Emojis are “a visual representation[s] of an emotion, object, or symbol” (Costa). Emojis are popular now through messaging apps and social media apps. Through this project, I wanted to interact with both visually and textually. I have read several modern novels that have used texting as a communication archetype and it is interesting to see how communication has changed. Even I have experienced change in communication, primarily going from longer emails to shorter text messages that occur more frequently over the day. I both love and hate seeing modern texting and vernacular in traditional books. There are some technology features that are hard to translate into traditional media. Even with text messages being featured in books, I have yet to see a picture of emoji, I usually see books use text emoticons such as ":)" or ":D" but not emojis. 
 <br> <br>
&nbsp;&nbsp; For my modification, I kept the lists’ names the same and ultimately tried to create very similar poems from the original The House of Dust. I deleted most, if not all, of the previous sentence fragments and attempted to replace them with the emoji-zed equivalent. There were also constraints using emojis using this approach. Another hurdle is that the Python library I had was not updated with the most recent emoji so some newer, commonly used emojis were not featured, even as I am modernizing The House of Dust mod, it is still not using the most update to date emoji library because they are constantly being developed. There are not emojis for every word, I struggled finding emojis that could be used as a material. Most of my materials are more abstract, like using snow and fire instead of stone and brick. This gives my poems a more impressionistic sense.
<br> <br>
&nbsp;&nbsp; While using the emojis, I found it hard to parse the meaning of text based solely on a small image, so I added additional textual context that helps the story telling more. Some stanzas utilize solely emojis while others have some more context. For example, in the materials list, I generally just used the emoji since the sentence structure for the first sentence can be understood without additional context. In some sentences, I used emojis to emphasize the content of the sentence. This is similar to how some people will use emojis as decorations for sentences versus how some people use emojis as substitutions for words. I really like how one can interpret a poem in multiple ways. By using visual representations, it allows a certain amount of freedom to interpret the poem.

Even though these are vary similar, they invoke different feelings, as starlight feels nostalgic, like a beam of starlight breaking through curtains while a house of stars has a magical feeling. " IN A GREEN, MOSSY TERRAIN 🚵" is from the original code but I added an emoji for emphasis. Due to the lack of lighting theme emojis, I had to think creatively about what emoji can be used as light. "USING JACK O LANTERNS 🎃" was one of my ideas. I imagine a mossy, starlit house up on a hill being softly illuminated using carved pumpkins as the faint smell of honey from the bees waft through the air.
